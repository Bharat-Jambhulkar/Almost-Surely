**Example 6.3.2**

```python
import numpy as np
def Xnw(n,w):   
    return np.exp(w)+w**n

def Xw(w):
    return np.exp(w)
```

```python
omegavec = np.linspace(0.999,0.9999,num=1000)
len(omegavec)
nvec = np.array([10,100,1000,5000,10000,50000])
```

```python
epsilon = np.array([0.1,0.5])
XnMat = np.zeros((len(omegavec),len(nvec)))
```

```python
for i in range(len(omegavec)):
    omega = omegavec[i]

    for j in range(len(nvec)):
        n = nvec[j]
        XnMat[i,j] = Xnw(n,omega)

XVec = []
for i in range(len(omegavec)):
    omega = omegavec[i]
    XVec.append(Xw(omega))
```

```python
nlab = ["n=" + str(n) for n in nvec]
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()

# Main line
ax.plot(omegavec, XVec, color='blue', linestyle='-', linewidth=2)

# Axis limits
ax.set_xlim(0.999, 1)
ax.set_ylim(XVec[0] - 0.7, np.max(XnMat) + 0.1)

# Remove default x ticks and set custom ones
ax.set_xticks(np.arange(0.999, 1.0001, 0.0005))

# Axis labels (math expression)
ax.set_xlabel(r'$\omega$')
ax.set_ylabel(r'$X(\omega)$')

# Title (math expression)
ax.set_title(r'$X_n(\omega) := e^{\omega} + \omega^n$')

# Text label near first curve
ax.text(0.999, XVec[0] - 0.05, r'$X(\omega)$', color='blue')

# Loop over columns of XnMat
for i in range(XnMat.shape[1]):
    ax.plot(omegavec, XnMat[:, i],
            linestyle='-',
            linewidth=2)
    
    ax.text(0.99995,
            XnMat[-1, i] - 0.01 * (-1)**i,
            nlab[i])

# Horizontal dashed lines (epsilon[0])
ax.axhline(XVec[0] + epsilon[0], color='blue', linestyle='--', linewidth=1.5)
ax.axhline(XVec[0] - epsilon[0], color='blue', linestyle='--', linewidth=1.5)

ax.text(0.9995, XVec[0] + 0.15, r'$X(\omega) + 0.1$', color='blue')
ax.text(0.9995, XVec[0] - 0.15, r'$X(\omega) - 0.1$', color='blue')

# Horizontal dotted lines (epsilon[1])
ax.axhline(XVec[0] - epsilon[1], color='green', linestyle=':', linewidth=1.5)
ax.axhline(XVec[0] + epsilon[1], color='green', linestyle=':', linewidth=1.5)

ax.text(0.9995, XVec[0] + 0.55, r'$X(\omega) + 0.5$', color='green')
ax.text(0.9995, XVec[0] - 0.55, r'$X(\omega) - 0.5$', color='green')

# Points (type="p" equivalent)
ax.plot(1, Xw(1), marker='x', color='blue')
ax.plot(1, Xnw(100, 1), marker='o', color='black')

plt.show()
```
