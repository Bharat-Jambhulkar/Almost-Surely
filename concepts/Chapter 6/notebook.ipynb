```python
import numpy as np
```

**Example 6.3.2**

```python
def Xnw(n,w):
    return np.exp(w)+w**n

def Xw(w):
    return np.exp(w)

omegavec = np.arange(0,1,0.01)
nvec = np.array([1,10,100])
XnMat = np.zeros((len(omegavec),len(nvec)))
```

```python
for i in range(len(omegavec)):
    omega = omegavec[i]
    
    for j in range(len(nvec)):
        n = nvec[j]
        XnMat[i,j] = Xnw(n,omega)

Xvec = []

for i in range(len(omegavec)):
    omega = omegavec[i]
    Xvec.append(Xw(omega))
```

```python
import matplotlib.pyplot as plt

plt.plot(omegavec,Xvec,color='blue',label=r'$X(\omega)$')
plt.plot(omegavec,XnMat[:,0],color='green',linestyle='--',label='n=1')
plt.plot(omegavec,XnMat[:,1],color='red',linestyle=':',label='n=10')
plt.plot(omegavec,XnMat[:,2],color='orange',linestyle='-.',label='n=100')

plt.title(r"$Xn(\omega) = exp(\omega) + \omega^n$")

plt.ylabel(r"$Xn(\omega) ~ or ~ X(\omega)$")

plt.legend()
```
